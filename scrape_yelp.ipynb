````python
cuisine_types = ['french','italian','japanese','korean','mexican','chinese','burgers','']
id_list = []
restaurant_summary = []
for types in cuisine_types:
  print('***** Searching for '+types+' Restaurants in New York *****')
  print("yelp_api.search_query(term='restaurants', categories=" + types + ", location='Mahanttan, New York'")
  cuisine_count = 0
  offset_count = 0
  try:
    # don't waste API calls, limit to 30*6 = 180 calls max  
    while cuisine_count < 1000 and offset_count < 30:    
        offset_count += 1
        # search 50 at a time, limited by api (I would have tried 1000)
        response = yelp_api.search_query(term='restaurants',categories= types , location='Mahanttan, New York', limit = 50, offset = 50*offset_count)
        for restaurant in response['businesses']:
          if restaurant['id'] not in id_list:
            id_list.append(restaurant['id'])
            restaurant_summary.append(restaurant)
            cuisine_count += 1
        # print 1 sample restaurant for sanity check
        print('\nRestaurant No.'+str(offset_count*50-49)+': ')
        pprint(response['businesses'][0])
  except YelpAPI.YelpAPIError as e:
    print(e)
    
    
  print('\n-------------------------------------------------------------------------\n')
````

````python
import json
with open('restaurant.json', 'w') as file:
    json.dump(restaurant_summary, file)
````
